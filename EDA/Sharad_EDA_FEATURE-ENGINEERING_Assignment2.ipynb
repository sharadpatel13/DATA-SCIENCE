```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset=pd.read_csv("C:\\Users\\SHARAD\\Downloads\\train.csv")
dataset.head()
```

```python
## Always remember there way always be a chance of data leakage so we need to split the data first and then apply feature
## Engineering
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(dataset,dataset['SalePrice'],test_size=0.1,random_state=0)
```

```python
X_train.shape, X_test.shape
```

```python
# MISSING VALUES

#Let us capture all the nan values
#First lets handle categorical features which are missing
feature_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes=='O']

for feature in feature_nan:
    print('{}: {}% missing values'.format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
## Replace missing value with a new label
def replace_feature(dataset,feature_nan):
    data=dataset.copy()
    data[feature_nan]=data[feature_nan].fillna('Missing')
    return data

dataset=replace_feature(dataset,feature_nan)
dataset.head()
```

```python
dataset[feature_nan].isnull().sum()
```

```python
#Now check for numerical variables that contains missing values
numerical_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes!='O']

#We will print the numerical nan variables and percentage of missing values
for feature in numerical_nan:
    print('{}: {}% missing value'.format(feature,np.around(dataset[feature].isnull().mean(),4)))
```

```python
#Replacing the numerical missing values
for feature in numerical_nan:
    ## We will replace by using median since there are outliers
    median_value=dataset[feature].median()
    
    ## create a new feature to capture nan values
    dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value)
    
print(dataset[numerical_nan].isnull().sum())
```

```python
dataset.head(20)
```

```python
## Temporal Variables (Date Time Variables)

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
       
    dataset[feature]=dataset['YrSold']-dataset[feature]
```

```python
dataset.head()
```

```python
dataset[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

```python
# Numerical Variables

#Since the numerical variables are skewed we will perform log normal distribution
```

```python
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
    dataset[feature]=np.log(dataset[feature])
```

```python
dataset.head()
```

```python
#  Handling Rare Categorical Feature

# We will remove categorical variables that are present less than 1% of the observations
```

```python
categorical_features=[feature for feature in dataset.columns if dataset[feature].dtype=='O']
```

```python
categorical_features
```

```python
for feature in categorical_features:
    temp=dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
    

dataset.head()
```

```python
for feature in categorical_features:
    labels_ordered=dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)

dataset.head(10)
```

```python
scaling_feature=[feature for feature in dataset.columns if feature not in ['Id','SalePerice'] ]
len(scaling_feature)
```

```python
#Feature Scaling
```

```python
feature_scale=[feature for feature in dataset.columns if feature not in ['Id','SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(dataset[feature_scale])
```

```python
scaler.transform(dataset[feature_scale])
```

```python
# transform the train and test set, and add on the Id and SalePrice variables
data = pd.concat([dataset[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(dataset[feature_scale]), columns=feature_scale)],
                    axis=1)
```

#### Feature Engineering Process

#### 1. Data Splitting:
The dataset is split into training and testing sets using the train_test_split function from sklearn.model_selection.
This step ensures that feature engineering is applied separately to training and testing data to prevent data leakage.

#### 2. Handling Missing Values:
Categorical Features:
Identified categorical features with missing values and replaced them with a new label, 'Missing', using a custom function replace_feature.
Numerical Features:
Identified numerical features with missing values and replaced them with the median since there are outliers.
Created new binary features to indicate whether a value was missing or not.

#### 3. Temporal Variables (Date Time Variables):
Transformed temporal variables ('YearBuilt', 'YearRemodAdd', 'GarageYrBlt') into the age of the property by subtracting from the 'YrSold' feature.
This transformation helps in capturing the age of the property, which might be more relevant than the actual year.

#### 4. Log Transformation for Numerical Variables:
Applied log transformation to skewed numerical variables ('LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice') to achieve a more normal distribution.
Log transformation helps in handling skewed distributions and reduces the impact of outliers.

#### 5. Handling Rare Categorical Features:
Identified categorical variables with rare categories (less than 1% of observations) and replaced them with the label 'Rare_var'.
This step helps in reducing the cardinality of categorical features and avoids overfitting.

#### 6. Ordinal Encoding for Categorical Features: 
For categorical features, ordinal encoding was performed based on the mean 'SalePrice' within each category.
This encoding preserves the ordinal relationship between categories while converting them into numerical values.

#### 7. Feature Scaling:
Used Min-Max scaling to scale the numerical features to a range between 0 and 1.
Scaling ensures that all features contribute equally to the model and prevents features with larger scales from dominating the model training process.
Challenges and Solutions:

Handling Missing Values:
Missing values were handled differently for categorical and numerical features to maintain data integrity.
A binary indicator was created for numerical features to capture missing values.
Rare Categorical Features:
Rare categories were replaced with a common label to reduce the dimensionality of categorical features without losing significant information.
Feature Scaling:
Min-Max scaling was chosen for its simplicity and ability to preserve the shape of the original distribution.
Comparison of Transformed and Original Datasets:

The transformed dataset (data) contains all the original features along with the engineered features.
The original dataset (dataset) retains its structure but with missing values filled, temporal variables transformed, numerical variables logged, and categorical variables encoded.
Both datasets can be compared to observe the effects of feature engineering on the data distribution and model performance.
